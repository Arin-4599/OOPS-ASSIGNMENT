```python
# 1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog
    #that overrides the speak() method to print "Bark!".

class Animal():
  def speak(self):
    print("every animal has different way of speaking")
class dog(Animal):
  def speak(self):
    print("dog speak bark!")
a1 = dog()
print(a1.speak())

# 2.  Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
#     from it and implement the area() method in both.
import abc
class shape():
  @abc.abstractmethod
  def calculate_area(self):
    pass
class circle(shape):
  def calculate_area(self):
    return "area of circle pi*r**2"
class rectangle(shape):
  def calculate_area(self):
    return "area of rectangle length * breadth"
a1 = circle()
print(a1.calculate_area())
a2 = rectangle()
print(a2.calculate_area())

# 3. implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
     #and further derive a class ElectricCar that adds a battery attribute.
class vechile():
  def vechile_info(self):
    print("vechile run on both battery and fuel")
class car(vechile):
  def vechile_info(self):
    print("car run on battery as well as fuel")
class Electriccar(car):
  def Electriccar_info(self):
    print("Electriccar run on only battery")
c1 = Electriccar()
print(c1.Electriccar_info())

# 4. . Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
     # Sparrow and Penguin that override the fly() method.
class bird():
  def fly(self):
    print("all birds can fly")
class sparrow(bird):
  def fly(self):
    print("sparrow can fly very fast")
class penguin(bird):
  def fly(self):
    print("penguin is the only bird that can not fly")
obj1 = sparrow()
obj2 = penguin()
class_obj = [obj1, obj2]
def parcer(class_obj):
  for i in class_obj:
    i.fly()
print(parcer(class_obj))

# 5.  Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
      #balance and methods to deposit, withdraw, and check balance.
class bank_acc():
  def __init__(self, balance):
    self.__balance = balance
  def deposite(self, amount):
    self.__balance = self.__balance + amount
  def withdraw(self, amount):
    if self.__balance >= amount:
      self.__balance = self.__balance - amount
      return True
    else:
      return False
  def get_balance(self):
    return self.__balance
a1 = bank_acc(10000)
print(a1.get_balance())
print(a1.withdraw(300))
a1.deposite(5000)
print(a1.get_balance())

# 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
    #and Piano that implement their own version of play().
class Instrument():
  def play(self):
    print("all instrument have different sound")
class guitar(Instrument):
  def play(self):
    print("it use string for creating sound")
class piano(Instrument):
  def play(self):
    print("it uses air for creating sound")
obj1 = guitar()
obj2 = piano()
class_obj = [obj1, obj2]
def playing(class_obj):
  for i in class_obj:
    i.play()
print(playing(class_obj))

# 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
     #method subtract_numbers() to subtract two numbers.
class MathOperations():
  def add_numbers(a, b):
    return a+b
  @staticmethod
  def subtract_numbers(a,b):
    return a-b
print(MathOperations.add_numbers(5,7))
print(MathOperations.subtract_numbers(10,8))

# 8. Implement a class Person with a class method to count the total number of persons created.
class person():
  count = 0
  def __init__(self, name):
    self.name = name
    person.count += 1
  @classmethod
  def total_persons(cls):
    return cls.count
p1 = person("arin")
p2 = person("harsh")
p3 = person("amit")
print("total person created:", person.total_persons())

# 9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
     #fraction as "numerator/denominator".
class fraction():
  def __init__(self, numerator, denominator):
    self.numerator = numerator
    self.denominator = denominator
  def __str__(self):
    return f"{self.numerator}/{self.denominator}"

f1 = fraction(3,4)
f2 = fraction(5,8)
print(f1)
print(f2)

# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
      #vectors.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2  # This calls v1.__add__(v2)

print("v1 =", v1)  # Output: (2, 3)
print("v2 =", v2)  # Output: (4, 5)
print("v1 + v2 =", v3)  # Output: (6, 8)








```
